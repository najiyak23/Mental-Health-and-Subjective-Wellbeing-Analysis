# Regression Analysis

```python
import statsmodels.api as sm

# Model 1
X1 = sm.add_constant(df_model[['PHQ-9_Score']])
y = df_model['wellbeing']
model1 = sm.OLS(y, X1).fit()
print(model1.summary())

# Model 2
X2 = sm.add_constant(df_model[['PHQ-9_Score','age','gender_1','ses_1']])
model2 = sm.OLS(y, X2).fit()
print(model2.summary())

# Model 3
X3 = sm.add_constant(df_model[['PHQ-9_Score','GAD-7_Score','PSS-10_Score','age','gender_1','ses_1']])
model3 = sm.OLS(y, X3).fit()
print(model3.summary())
```

• Depression is a strong and significant negative predictor of wellbeing.

• Depression alone explains about 25% of the variation in wellbeing.

• Adding age, gender, and SES does not improve the model meaningfully.

• Age, gender, and SES are not significant predictors of wellbeing.

• When anxiety and stress are included, all three (depression, anxiety, stress) become significant predictors.

• Anxiety shows the strongest negative effect on wellbeing, followed by stress, then depression.

• The effect of depression reduces substantially after adding anxiety and stress.

• This reduction indicates shared variance and multicollinearity among mental health variables.

• The full model explains about 33% of the variation in wellbeing.

• Demographic variables remain non-significant even in the full model.

## Quantile Regression

```python
import statsmodels.api as sm

# Define predictors
X = df_model[['PHQ-9_Score','GAD-7_Score','PSS-10_Score','age','gender_1','ses_1']]
X = sm.add_constant(X)

# Dependent variable
y = df_model['wellbeing']
```

```python
model_q50 = sm.QuantReg(y, X).fit(q=0.5)
print(model_q50.summary())
```

```python
quantiles = [0.25, 0.5, 0.75]
models = {}

for q in quantiles:
    model = sm.QuantReg(y, X).fit(q=q)
    models[q] = model
    print(f"\nQuantile: {q}")
    print(model.summary())
```

**Insights**

• Depression, anxiety, and stress are significant negative predictors of median wellbeing.

• Anxiety shows the strongest negative association, followed by depression and stress.

• The direction and significance of effects are consistent with OLS results, confirming robustness.

• Demographic variables (age, gender, SES) remain non-significant.

• Quantile regression results indicate that findings are stable even after accounting for outliers and non-constant variance.

• This suggests that the relationship between mental health and wellbeing is consistent across different modeling approaches.

```python
import pandas as pd

coef_df = pd.DataFrame({
    q: models[q].params for q in quantiles
})

print(coef_df)
```

**Insights**

• Mental health factors (depression, anxiety, stress) reduce wellbeing across all levels.

• Anxiety and stress have the strongest effect at lower wellbeing (0.25 quantile), indicating greater impact among individuals with poorer wellbeing.

• Their effects weaken at higher wellbeing levels (0.75 quantile), suggesting greater resilience among individuals with higher wellbeing.

• Depression shows a relatively consistent effect, strongest around the median (0.50 quantile).

• Age, gender, and SES do not show a meaningful impact on wellbeing.
